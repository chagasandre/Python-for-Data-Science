```python
# Importando  bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import seaborn as sb
```

```python
#Importando os arquivos
cadastros_df = pd.read_csv('cadastro.csv')
cadastros_df['Quantidade'] = 1
cadastros_df['idade'] = 2020 - cadastros_df['ano_nascimento']

# Exibindo a média
print("Média de idades: ",round(cadastros_df['idade'].sum() / cadastros_df['Quantidade'].sum(),2))
```

```python
#Exibindo o gráfico de mulheres e homens
cadastros_df[['sexo','Quantidade']].groupby('sexo').sum().plot(kind='barh')
```

```python
#Renomeando para coletar a informação correta
cadastros_df["escolaridade"].loc[cadastros_df["escolaridade"] == "ensino médio"] = "Ensino Médio"
cadastros_df["escolaridade"].loc[cadastros_df["escolaridade"] == "ensino Médio"] = "Ensino Médio"
cadastros_df["escolaridade"].loc[cadastros_df["escolaridade"] == "ensino superior"] = "Ensino Superior"
cadastros_df["escolaridade"].loc[cadastros_df["escolaridade"] == "Ensimo Tecnico"] = "Ensino Técnico"
cadastros_df[['escolaridade','Quantidade']].groupby('escolaridade').sum().plot(kind="barh")
```

```python
#Importando os arquivos
salarios_df = pd.read_csv('salario_profissao.csv')

#Coletando informações das tabelas
df_salarios_profissoes = cadastro_df.merge(salario_df, left_on='profissão', right_on='Profissões')
df_salarios_profissoes['salario'] = df_salarios_profissoes["salario até 3 anos"].loc[df_salarios_profissoes["experiencia profissional"] <= 3]
df_salarios_profissoes['salario'] = df_salarios_profissoes['salario'].fillna(df_salarios_profissoes["salaior +3 anos"])
df_salarios_profissoes.groupby('salario').sum().plot(kind="barh")
print("Média Salarial: ",round(df_salarios_profissoes['salario'].sum() / cadastros_df['Quantidade'].sum(),2))
```

```python
#Exibindo o maior salario
maior_salario = max(df_salarios_profissoes["salario"])
df_salarios_profissoes.loc[df_salarios_profissoes["salario"] == maior_salario]
print("O maior salário é:", maior_salario)
```

```python
#Exibindo o menor salario
menor_salario = min(df_salarios_profissoes["salario"].loc[df_salarios_profissoes["salario"] > 0])
df_salario_profissao.loc[df_salarios_profissoes["salario"] == menor_salario]
print("O menor salário é:", menor_salario)
```
