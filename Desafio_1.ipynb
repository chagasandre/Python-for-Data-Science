```python
#biblioteca para formatar data

ref_arquivo = open("/Users/chaga/Documents/Impacta/Python/conversa do whatsapp.txt","r", encoding="utf-8")
linha = ref_arquivo.readline()

list_datas = []
list_erros_formatacao = []
list_pessoas = []
list_msgs = []
list_horas = []

while linha:
    linha = ref_arquivo.readline()
    
    try:
        data_hora = linha[:16]
        if(len(data_hora.split()) >= 2):
            list_datas.append(data_hora.split()[0])
            list_horas.append(data_hora.split()[1])
            for data in list_datas:
                if((len(data) < 10 or len(data)> 10) and not data.find('-')):
                    list_datas.remove(data)
            for hora in list_horas:
                if(len(hora) < 5 or len(hora)> 5):
                    list_horas.remove(hora)
        msg = linha[19:]
        if(len(msg.split(':')) >= 2):
            list_pessoas.append(msg.split(':')[0])
            list_msgs.append(msg.split(':')[1])
            for pessoa in list_pessoas:
                if(len(pessoa) > 20):
                    list_pessoas.remove(pessoa)
        
    except ValueError:
        list_erros_formatacao.append('Erros de formata√ß√£o de data')





ref_arquivo.close()        
        
print(list_datas)
print(list_horas)
print(list_pessoas)
print(list_msgs)

```

```python
def contador_palavras(msgs):
    palavras = []
    for frases in msgs:
        for palavra in frases.split():
            palavras.append(palavra)
    return Counter(palavras)

palavras_dict = dict(contador_palavras(list_msgs))

palavras_df = pd.DataFrame(palavras_dict.items(), columns=['Palavra', 'Frequencia'])
palavras_df = palavras_df.sort_values(by=['Frequencia'], ascending=False)
palavras_df.head(10).plot(kind='bar', x = 'Palavra', y='Frequencia', title='Palavras mais faladas')
```
